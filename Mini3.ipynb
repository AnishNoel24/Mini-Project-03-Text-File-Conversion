```python
import time

# Function to convert file to upper case
def convert_to_upper_case(file_path):
    with open(file_path, 'r') as file:
        content = file.read().upper()
    with open(file_path, 'w') as file:
        file.write(content)

# File sizes to test
file_sizes = [200, 400, 600, 800, 1000]  # MB

# Perform conversion for each file size
for size in file_sizes:
    file_path = f"input_{size}mb.txt"
    # Create text file of given size (filled with random characters)
    with open(file_path, 'w') as file:
        file.write(''.join(['a' for _ in range(size * 1024 * 1024)]))  # MB to bytes conversion

    # Measure execution time
    start_time = time.time()
    convert_to_upper_case(file_path)
    end_time = time.time()

    # Calculate and print execution time
    execution_time = end_time - start_time
    print(f"File size: {size}MB, Execution time: {execution_time} seconds")

```

```python
import matplotlib.pyplot as plt

file_sizes = [200, 400, 600, 800, 1000]  # MB

c_plus_plus_times = [3.09, 6.29, 9.54, 13.35, 16.33]
c_times = [3.62, 6.68, 6.29, 9.92, 13.13]
python_times = [1.04, 5.19, 9.21, 6.32, 8.53]
java_times = [4.47, 9.59, 14.11, 21.02, 25.99]
r_times = [4.85, 15.13, 17.97, 19.65, 22.35]

plt.plot(file_sizes, c_plus_plus_times, marker='o', label='C++')
plt.plot(file_sizes, c_times, marker='o', label='C')
plt.plot(file_sizes, python_times, marker='o', label='Python')
plt.plot(file_sizes, java_times, marker='o', label='Java')
plt.plot(file_sizes, r_times, marker='o', label='R')

plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')
plt.title('Performance Comparison - File Conversion to Upper Case')
plt.legend()
plt.grid(True)
plt.show()

```
